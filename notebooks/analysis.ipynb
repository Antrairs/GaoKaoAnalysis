```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d2021 = pd.read_csv("../data/2021.csv")
d2022 = pd.read_csv("../data/2022.csv")
d2023 = pd.read_csv("../data/2023.csv")
```

```python
class Statistics:
    def __init__(self, year, score, num):
        self.year = year
        self.total =num.sum()
        wig = num/self.total
        self.mu = (wig*score).sum()
        self.sigma = (wig*((score-self.mu)**2)).sum()**0.5
```

```python
def normal_dis(x, mu, sigma):
    return 0.3989422804014327 / sigma * np.exp( - (x-mu)**2 / (2*sigma**2) )
```

```python
phy2021 = Statistics(2021, score=d2021["score"], num=d2021["physics"])
phy2022 = Statistics(2022, score=d2022["score"], num=d2022["physics"])
phy2023 = Statistics(2023, score=d2023["score"], num=d2023["physics"])
```

```python

plt.scatter(x=d2021["score"], y=d2021["physics"],s=10, c="r")
plt.scatter(x=d2022["score"], y=d2022["physics"],s=10, c="g")
plt.scatter(x=d2023["score"], y=d2023["physics"],s=10, c="b")
plt.show()
```

```python
plt.scatter(x=d2021["score"], y=d2021["physics"],s=10, c="r")
plt.plot(d2021["score"], normal_dis(d2021["score"], phy2021.mu, phy2021.sigma) * phy2021.total, c = "r")
plt.title(f"${phy2021.year}year  (\\mu={phy2021.mu:.2f}, \\sigma={phy2021.sigma:.2f}) $")
plt.show()
```

```python
plt.scatter(x=d2022["score"], y=d2022["physics"],s=10, c="g")
plt.plot(d2022["score"], normal_dis(d2022["score"], phy2022.mu, phy2022.sigma) * phy2022.total, c = "g" )
plt.title(f"${phy2022.year}year (\\mu={phy2022.mu:.2f}, \\sigma={phy2022.sigma:.2f})$")
plt.show()
```

```python
plt.scatter(x=d2023["score"], y=d2023["physics"],s=10, c="b")
plt.plot(d2023["score"], normal_dis(d2023["score"], phy2023.mu, phy2023.sigma) * phy2023.total, c = "b")
plt.title(f"${phy2023.year}year (\\mu={phy2023.mu:.2f}, \\sigma={phy2023.sigma:.2f}) $")
plt.show()
```
